```python
import os
import pandas as pd

results_dir = '/content/drive/MyDrive/TOPO_COMPARE/results/final'

desired_datasets = ['mnist', 'fmnist', 'cifar10', 'pbmc3k', 'paul15', 'earth', 'mammoth', 'spheres', 'swiss_roll', 'concentric_spheres']
desired_models = ['mmae', 'ggae', 'geomae', 'vanilla', 'topoae', 'rtdae', 'spae']
desired_dims = [2, 16, 64, 128]

found_results = set()
csv_files = []

print(f"Searching for results in: {results_dir}")

# Recursively list all CSV files in the directory and its subdirectories
try:
    for root, _, files in os.walk(results_dir):
        for f_name in files:
            if f_name.endswith('.csv') and f_name.startswith('final_results_'):
                csv_files.append(os.path.join(root, f_name))

    if not csv_files:
        print("No relevant CSV files found in the directory or its subdirectories.")

    for file_path in csv_files:
        f_name = os.path.basename(file_path)

        # Extract dataset name from filename (e.g., final_results_fmnist_...csv -> fmnist)
        parts = f_name.replace('final_results_', '').split('_')
        dataset_name = parts[0]

        if dataset_name not in desired_datasets:
            print(f"Skipping file {f_name}: Dataset '{dataset_name}' is not in the desired list.")
            continue

        try:
            df = pd.read_csv(file_path)

            # Normalize column names for easier access
            df.columns = df.columns.str.lower()

            # Ensure required columns exist, checking for 'dim' or 'latent_dim'
            has_model_col = 'model' in df.columns
            has_dim_col = 'dim' in df.columns or 'latent_dim' in df.columns
            has_wasserstein_col = 'wasserstein_h0' in df.columns

            if not (has_model_col and has_dim_col and has_wasserstein_col):
                print(f"Skipping file {f_name}: Missing one or more of 'model', 'dim'/'latent_dim', or 'wasserstein_h0' columns.")
                continue

            for _, row in df.iterrows():
                model = str(row['model']).lower().strip()

                # Get dimension from 'dim' or 'latent_dim'
                dim = None
                if 'dim' in row and pd.notna(row['dim']):
                    dim = int(row['dim'])
                elif 'latent_dim' in row and pd.notna(row['latent_dim']):
                    dim = int(row['latent_dim'])

                if dim is None:
                    continue # Skip if no valid dimension found

                if model in desired_models and dim in desired_dims:
                    found_results.add((dataset_name, model, dim))

        except Exception as e:
            print(f"Error reading {f_name}: {e}")

except FileNotFoundError:
    print(f"Error: Directory not found at {results_dir}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")

print("\n--- Synthesis of Results ---")

summary = {} # Store results for better display

for dataset in desired_datasets:
    summary[dataset] = {}
    for model in desired_models:
        summary[dataset][model] = {}
        for dim in desired_dims:
            status = "Found" if (dataset, model, dim) in found_results else "Missing"
            summary[dataset][model][dim] = status

# Print a formatted table
header = "{:>18}".format("Dataset") + "|" + "{:>10}".format("Model") + "|" + " ".join([f"D={d:<3}" for d in desired_dims])
print(header)
print("-" * len(header))

for dataset in desired_datasets:
    for model in desired_models:
        row_str = "{:>18}".format(dataset) + "|" + "{:>10}".format(model) + "|"
        for dim in desired_dims:
            status_char = "✓" if summary[dataset][model][dim] == "Found" else "X"
            row_str += f"  {status_char}  "
        print(row_str)
```
